### 4.4.5. Making Departments Variable

```python
df_deps
```

```python
# Grouping orders by user_id
grouped_orders = df_merge_deps.groupby('user_id')
```

```python
# Creating Labels Variable
result = []

for user_id, user_orders in df_merge_deps.groupby('user_id')['department']:
    user_label = 'No label'  # Default label for each user
    has_meat_seafood = user_orders.str.contains('meat seafood').any()
    has_dairy_eggs = user_orders.str.contains('dairy eggs').any()
    
    if has_meat_seafood:
        user_label = 'Carnivore'
    elif user_orders.str.contains('babies').any() and user_orders.str.contains('pets').any():
        if not has_dairy_eggs:
            user_label = 'Vegetarian/Vegan that has baby and pet'
        else:
            user_label = 'Has baby and pet'
    elif user_orders.str.contains('babies').any():
        user_label = 'Has baby'
    elif user_orders.str.contains('pets').any():
        user_label = 'Has pet'
    
    if not has_meat_seafood and not has_dairy_eggs and user_label == 'No label':
        user_label = 'Vegan/Vegetarian'
    
    result.extend([user_label] * len(user_orders))
```

```python
result
```

```python
# Renaming result
df_merge_deps['user_label'] = result
```
