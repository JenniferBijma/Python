### Creating Loyalty Flag

```python
# Defining Max Order
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
# Ensuring 100 rows can be displayed
pd.options.display.max_rows = None
```

```python
# Checking the new Column
ords_prods_merge.head(100)
```

#### Creating Loyalty Flags

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking the New Flags
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']}).round(0)
```

### Creating Spender Flags

```python
ords_prods_merge.loc[ords_prods_merge['prices'] >= 10, 'spender_type'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] < 10, 'spender_type'] = 'Low spender'
```

```python
# Checking the Spender Type Column
ords_prods_merge['spender_type'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'prices', 'loyalty_flag', 'spender_type']].head(100)
```

### Creating Order Frequency Flags

```python
ords_prods_merge.loc[ords_prods_merge['days_since_prior_order'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['days_since_prior_order'] > 10) & (ords_prods_merge['days_since_prior_order'] <= 20), 'order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['days_since_prior_order'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Checking the Order Frequency column
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'prices', 'loyalty_flag', 'spender_type', 'order_frequency']].head(100)
```
